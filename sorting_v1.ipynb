```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.show()
```

```python
y, h = 0, 100
x, w = 0, 100
crop_img = img_rgb[y:y+h, x:x+w]

plt.imshow(crop_img)
plt.show()
```

```python
x_step = 8
y_step = 8
img_list = []
k = 0
for i in range(0, img_rgb.shape[0], y_step):       # y
    for j in range(0, img_rgb.shape[1], x_step):   # x
        # print(i, '\t', i + y_step -1, '\t', j, '\t', j + x_step -1)
        crop_img = img_rgb[i : i + y_step, j : j + x_step]
        img_list.append([k, crop_img])
        k += 1    

len(img_list)   # (512*512)/(8*8)
```

```python
k = 0
v_list = []
for i in range(int(img_rgb.shape[0]/y_step)):        # y 
    h_list = []
    for j in range(int(img_rgb.shape[1]/x_step)):    # x
        # print(k, end = ' ')
        h_list.append(img_list[k][1])
        k += 1
    # print()
    img_h = cv2.hconcat(h_list) 
    v_list.append(img_h)

reformed_img = cv2.vconcat(v_list)

plt.imshow(reformed_img)
plt.show()
```

```python
def bubble_sort(list1):  
    # Outer loop for traverse the entire list  
    for i in range(0,len(list1)-1):  
        for j in range(len(list1)-1):  
            if(list1[j]>list1[j+1]):  
                temp = list1[j]  
                list1[j] = list1[j+1]  
                list1[j+1] = temp  
    return list1  
  
list1 = [60, 59, 25, 63, 33, 32, 21, 11, 37, 56, 45, 4, 57, 44, 61, 51, 26, 22, 29, 39, 41, 46, 18, 50, 10, 17, 13, 1, 52, 36, 2, 14, 16, 34, 20, 49, 43, 42, 12, 23, 54, 15, 3, 35, 24, 48, 5, 38, 9, 40, 19, 58, 7, 28, 31, 53, 8, 30, 27, 62, 0, 6, 47, 55]
print("The unsorted list is: ", list1)  
# Calling the bubble sort function  
print("The sorted list is: ", bubble_sort(list1)) 
```

```python
def selectionSort(array, size):
    
    for ind in range(size):
        min_index = ind
 
        for j in range(ind + 1, size):
            # select the minimum element in every iteration
            if array[j] < array[min_index]:
                min_index = j
         # swapping the elements to sort the array
        (array[ind], array[min_index]) = (array[min_index], array[ind])
        print(array)
 
arr = [61, 62, 58, 63, 57, 55, 54, 52, 53, 51, 50, 60, 59, 49, 48, 47, 56, 46, 45, 44, 43, 42, 41, 40, 39, 38, 37, 36, 35, 34, 33, 32, 30, 31, 28, 29, 27, 26, 25, 24, 23, 22, 21, 20, 19, 17, 18, 16, 15, 14, 13, 12, 11, 10, 8, 9, 7, 6, 5, 4, 2, 0, 1, 3]
size = len(arr)
selectionSort(arr, size)
print('The array after sorting in Ascending Order by selection sort is:')
print(arr)
```
